```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def naive_par_time(n, t):
    return n / t + t

def par_w_scan_time(n, t):
    return n / t + np.log2(t)

def serial_time(n, t):
    return n

def scan_naive_time(n, t):
    return ((n - 1) / (2 * (t - 1))) * np.log2(t)
```

```python
n = 10000
ts = np.arange(1, n, 1)

plt.figure()
plt.plot(ts, [n] * len(ts), label="serial")
plt.plot(ts, naive_par_time(n, ts), label="par")
# plt.plot(ts, scan_naive_time(n, ts), label="scan")
plt.plot(ts, par_w_scan_time(n, ts), label="scan + par")
plt.legend()
plt.yscale('log')
plt.xscale('log')
plt.ylabel('PARALLEL TIME CONSUMPTION')
plt.xlabel('NUM THREADS')
```

```python
naive_par_time(n, 128) / par_w_scan_time(n, 128)
```
